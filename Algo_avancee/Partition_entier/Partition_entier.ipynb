<h1> <span style="font-size:1.2em; color:orange"> 1. Définitions et notations mathématiques </span> </h1>


On souhaite dénombrer et expliciter les décompositions d'entiers en somme d'entiers.
On appelle partition de l'entier $n∈\mathbb{N}$ un élément de l'ensemble:

$D_n=\left\{ \left( a_1;a_2;...;a_m   \right) / m∈\mathbb{N}^* ; a_{i}∈\mathbb{N}^*; 0<a_1≤a_2≤...≤a_m ; n= \sum_{i=1}^{m}{a_i } \right\} $ si $n∈\mathbb{N}^*$


$D_0=\left\{ ∅ \right\}$  ($0$ a une unique partition, qui est vide)



Plus précisément, un élément $\left( a_1;a_2;...;a_m \right)∈D_n$ est appelé partition de $n$ de longueur $m$, ou encore $m$-partition de $n$. 

On notera $P(n,k)$ l'ensemble des partitions de $n$ de longueurs inférieures ou égales à $k$, et $p(n,k)$ le cardinal de cet ensemble.

Par exemple, pour $n=5$, il y a:
- une partition de longueur 1 qui est $5=5$
- deux partitions de longueur 2 qui sont $5=1+4$ et $5=2+3$
- deux partitions de longueur 3 qui sont $5=1+2+2$ et $5=1+1+3$
- une partition de longueur 4 qui est $5=1+1+1+2$
- une partition de longueur 5 qui est $5=1+1+1+1+1$

On a donc $P(5,3)=\left\{ \left(5\right);\left(1;4\right);\left(2;3\right);             \left(1;2;2\right);\left(1;1;3\right)\right\}$ et $p(5,3)=5$.

<h1> <span style="font-size:1.2em; color:orange"> 2. Dénombrement des partitions d'un entier </span> </h1>

<span id="Titre21">
    
<h2> <span style="font-size:0.9em; color:orange"> 2.1. Formule de récurrence </span> </h2>    



- L'entier $0$ possède une unique partition, qui est de longueur nulle, ne possédant aucun élément et correspondant à une somme vide. Ainsi, on a $∀k∈\mathbb{N}$; $p(0,k)=1$. 


- Par convention, la somme d'un ensemble vide d'éléments est nulle. Ainsi, $n∈\mathbb{N}^*$ ne possède aucune partition de longueur nulle. On a donc: $∀n∈\mathbb{N}^*$; $p(n,0)=0$. 


- Supposons $k≤n$. 

    On associe alors à chaque $\left( a_1;a_2;...;a_m \right)∈P(n-k,k)$ de façon unique un $k$-uplet $\left( b_1;b_2;...;b_k \right)=\left( 1;1;...;1;a_1+1;a_2+1;...;a_m+1 \right)$ composé de $k-m$ termes égaux à $1$. 

    On a $0<1≤1≤...≤1≤a_1+1≤a_2+1≤...≤a_m+1$ donc $0<b_1≤b_2≤...≤b_k$

    et d'autre part $\sum_{i=1}^{k}{b_i }=(k-m)\times1+\sum_{i=1}^{m}{(a_i+1) }=k-m+\sum_{i=1}^{m}{a_i }+m=k+n-k=N$ 

    $\left( b_1;b_2;...;b_k \right)$ constitue donc une $k$-partition de $n$. 

    Réciproquement, à toute $k$-partition $\left( b_1;b_2;...;b_k \right)$ de $n$ on fait correspondre un unique élément $\left( b_1-1;b_2-1;...;b_k-1 \right)$ qui, en supprimant les éventuelles valeurs nulles, donne un élément $\left( a_1;a_2;...;a_m \right)∈P(n-k,k)$. Ceci prouve que l'ensemble des $k$-partitions de $n$ est en bijection avec $P(n-k,k)$. En considérant la décomposition en union disjointe $P(n,k)=P(n,k-1)∪${ensemble des $k$-partitions de $n$} et en passant aux cardinaux de ces ensembles, on obtient $p(n,k)=p(n,k-1)+p(n-k,k)$.
 

- Si $m>n$ alors il ne peut exister de $m$-partition de $n$. En effet, une telle partition $\left( a_1;a_2;...;a_m \right)$ vérifierait $0<a_1≤a_2≤...≤a_m$ et on aurait alors $\sum_{i=1}^{m}{a_i }≥\sum_{i=1}^{m}{1 }=m>n$, ce qui contredit $n=\sum_{i=1}^{m}{a_i }$. 
    
    On en déduit donc que $∀k>n$ ; $p(n,k)=p(n,n)$ (car il n'existe aucune $m$-partition de $n$ pour $n<m≤k$).


Bilan: Finalement, on a obtenu les 4 relations suivantes:
- $∀k∈\mathbb{N}$; $p(0,k)=1$ 
- $∀n∈\mathbb{N}^*$; $p(n,0)=0$
- $∀1≤k≤n$ ; $p(n,k)=p(n,k-1)+p(n-k,k)$
- $∀k>n$ ; $p(n,k)=p(n,n)$

<h2> <span style="font-size:0.7em; color:orange"> 2.2.1. Algorithme récursif direct </span> </h2> 


Les relations vues en II.1. permettent d'écrire un algorithme récursif de calcul des $p(n,k)$.

```python
def p_rec(n,k):
    """
    Algorithme récursif de calcul des p(n,k)
    """
    if n==0: return 1
    if k==0: return 0
    if n>=k: return p_rec(n,k-1)+p_rec(n-k,k)
    return p_rec(n,n)


def Test_element(ptest,n0,k0,txt="p"):
    """
    Procédure de test: calcul d'un coefficient ptest(n0,k0) avec une fonction ptest
    """
    print("Test de calcul avec la fonction ",ptest.__name__," : \n",txt,"(",n0,",",k0,")=",ptest(n0,k0),sep='')
    return None

Test_element(p_rec,30,18)
```

On peut tester l'efficacité de la fonction p_rec, en estimant le temps de calcul de chacun des termes $p(n,k)$.

```python
from time import clock
import numpy as np

def Test_temps_calcul(p,n,k):
    """
    Pour une fonction p de calcul des p(n,k),
    renvoie le coefficient p(n,k) et le temps de calcul de ce coefficient
    """
    t=clock() ; x=p(n,k) ; t=clock()-t
    return x,t  

def Test_efficacite(p,N):
    """
    Pour une fonction p de calcul des p(n,k),
    renvoie une matrice L contenant les p(n,k) avec 0<=n<=N et 0<=k<=N
    et une matrice T contenant les temps de calculs correspondants    
    """
    L=[[None for k in range(N+1)] for n in range(N+1)]
    T=[[None for k in range(N+1)] for n in range(N+1)]
    for n in range(N+1):
        for k in range(N+1):
            L[n][k],T[n][k]=Test_temps_calcul(p,n,k)
    return L,T

def Test_efficacite_affichage(p,N,renvoi=False):
    """
    Affiche les résultats de la fonction Test_efficacite
    L'argument renvoi permet:
      si True: de renvoyer le temps de calcul total (en s)
      si False: d'afficher l'ensemble des temps de calculs
    """ 
    L,T=Test_efficacite(p,N)
    temps_total=sum(sum([T[n][k] for k in range(n+1)]) for n in range(N+1))
    
    if renvoi: return temps_total 
    
    print("\nFonction utilisée:",p.__name__)
    print("\n\nLes p(n,k) calculés pour 0<=k<=",N," et 0<=n<=",N," sont:\n\n",sep='')
         
    print(np.array(L))
    
    print("\n\n et les temps de calculs correspondant (en s) sont: \n\n")
    print(np.array(T))
    
    print("\n\nLe temps de calcul total (en s) est :", temps_total)
          
    return None 


Test_efficacite_affichage(p_rec,40)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator


def Graphique(T,N,legendex='n',legendey='k',legendez='t (en s)'):
    """
    Affiche un graphique 3d permettant de visualiser les valeurs d'une matrice T
    n en abscisses, k en ordonnées, mat[n][k] en côte
    """
      
    max_t=max([max([T[n][k] for k in range(N+1)]) for n in range(N+1)])
    
    fig = plt.figure()
    axes = fig.add_subplot(111, projection='3d')

    plt.xticks(fontsize=12); plt.yticks(fontsize=12)
     
    x,y,z,couleur=[],[],[],[]

    for n in range(N+1):
        for k in range(N+1):    
            x.append(n)
            y.append(k)
            z.append(T[n][k])
            coeff_couleur=T[n][k]/max_t
            couleur.append((coeff_couleur,0.5-0.5*coeff_couleur,1-coeff_couleur))

    axes.scatter(x,y,z,color=couleur) 

    axes.set_xlim(0, N) ; axes.set_xlabel(legendex) ; axes.xaxis.set_major_locator(MultipleLocator(N//5))
    axes.set_ylim(0, N) ; axes.set_ylabel(legendey) ; axes.yaxis.set_major_locator(MultipleLocator(N//5))
    axes.set_zlim(0, 1.1*max_t) ; axes.set_zlabel(legendez) ; axes.zaxis.set_major_locator(MultipleLocator(max_t))
    
    plt.show()
    return None
    
def Graphique_temps_fonction(p,N,texte="\nGraphique donnant le temps de calcul de chaque p(n,k) pour la fonction",legendex='n',legendey='k',legendez='t (en s)'):
    """
    Affiche un graphique 3d permettant de visualiser les temps de calcul des p(n,k) avec la fonction p choisie
    n en abscisses, k en ordonnées, T[n][k] en côte
    Renvoie la matrice des temps de calcul
    """
    L,T=Test_efficacite(p,N)
    print(texte,p.__name__)
    Graphique(T,N,legendex,legendey,legendez)
    return T  
```

<h2> <span style="font-size:0.7em; color:orange"> 2.2.2. Algorithme récursif avec mémorisation </span> </h2> 


Pour éviter que les calculs soient effectués plusieurs fois dans la mise en oeuvre de l'algorithme récursif, on peut l'adapter en algorithme récursif avec une matrice de mémorisation.

```python
def p_rec_mem(n,k,mat=None):
    """
    Algorithme récursif de calcul des p(n,k) avec mémorisation
    """
    if not mat: mat=[[-1 for j in range(k+1)] for m in range(n+1)]
     
    if mat[n][k]<0:
        
        if n==0: mat[n][k]=1
        elif k==0: mat[n][k]=0
        elif n>=k: mat[n][k]=p_rec_mem(n,k-1,mat)+p_rec_mem(n-k,k,mat)
        else: mat[n][k]=p_rec_mem(n,n,mat)
    
    return mat[n][k]



Test_element(p_rec_mem,30,18)
```

<h2> <span style="font-size:0.7em; color:orange"> 2.2.3. Comparaison des algorithmes </span> </h2> 
  

Entre l'algorithme récursif direct et l'algorithme récursif avec mémorisation, on observe un gain en temps d'exécution. 

Le graphique ci-dessous permet de visualiser les coefficients multiplicateurs exponentiels de gains en vitesse, calculés avec la formule $m(n,k)=ln \left( \frac{T[p-rec](n,k)}{T[p-rec-mem](n,k)} \right)$ où $T[p-rec](n,k)$ et $T[p-rec-mem](n,k)$ sont les temps de calcul de $p(n,k)$ respectivement avec les fonctions p_rec et p_rec_mem.

Ainsi, on a $T[p-rec-mem](n,k)= e^{-m(n,k)} \times T[p-rec](n,k) $.

NB1: Le fait de choisir un coefficient qui soit exponentiel permet d'obtenir un graphique plus lisible.

NB2: Sur le graphique, les éventuelles valeurs $m(n,k)$ négatives sont remplacées par 0.

```python
mat_coeff=list(np.maximum(np.log(np.array(Temps_rec)/np.array(Temps_rec_mem)),0))

plt.rcParams['figure.figsize'] = [20, 10] ; plt.rcParams.update({'font.size': 14})
Graphique(mat_coeff,len(mat_coeff)-1,legendez="m(n,k)")
```

Malgré l'amélioration apportée à l'algorithme récursif, les tests effectués pour déterminer à chaque étape s'il faut compléter la matrice des résultats ne sont pas négligeables en temps d'exécution. Pour éviter ces tests, nous allons déterminer l'ordre optimal dans lequel remplir cette matrice, et en déduire un algorithme itératif effectuant ce remplissage.

Le graphique construit ci-dessous permet de matérialiser les dépendances entre les différents termes $p(n,k)$.


```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
plt.rcParams['figure.figsize'] = [20, 10]

def Graphique_Lien_termes(N):

    #palette de couleurs pour les points (initialisation à 0, à 1 et -1,-2 pour autres)
    palette_point={1:'deeppink', 0:'darkviolet', -1:'slateblue', -2:'tomato' }
    #propriétés des points par code couleur
    propriete_point={1:'$n=0$', 0:'$n \geq 1$ et $k=0$', -1:'$n \geq k \geq 1$', -2:'$n < k$' }

    #palette de couleurs pour les vecteurs
    palette_vecteur={0:'olive', 1:'darkgoldenrod', 2:'darkkhaki'}
    #propriétés des vecteurs par code couleur
    propriete_vecteur={0:'$p(n,k)$'+' avec '+'$p(n,k-1)$', 1:'$p(n,k)$'+' avec '+'$p(n-k,k)$', 2:'$p(n,k)$'+' avec '+'$p(n,n)$'}

    axes = plt.axes()
    axes.set_xlabel('n',fontsize=22)
    axes.set_ylabel('k',fontsize=22)

    plt.xlim(-0.25,N+0.25)
    plt.ylim(-0.25,N+0.25)

    x_point=[]; y_point=[]; couleur_point=[] ; 
    x_debvecteur=[]; y_debvecteur=[]; dx_vecteur=[]; dy_vecteur=[]; couleur_vecteur=[]

    marge=0.09 #marge pour une meilleure distinction entre points et vecteurs 
    decalage1=[0 for k in range(N+1)] #décalage pour une meilleure visualisation des vecteurs (non superposés)
    
    for n in range(N+1):
            decalage2=0 #décalage pour une meilleure visualisation des vecteurs (non superposés)               
            for k in range(N+1):    
                x_point.append(n)
                y_point.append(k)
                if n==0:
                    couleur_point.append(palette_point[1])
                elif k==0:
                    couleur_point.append(palette_point[0])
                elif n>=k: 
                    couleur_point.append(palette_point[-1])

                    #vecteurs pour p_rec(n,k-1)
                    x_debvecteur.append(n); y_debvecteur.append(k-1+marge)
                    dx_vecteur.append(0); dy_vecteur.append(1-2*marge)
                    couleur_vecteur.append(palette_vecteur[0])

                    #vecteurs pour p_rec(n-k,k)
                    x_debvecteur.append(n-k+marge); y_debvecteur.append(k+decalage1[k])
                    dx_vecteur.append(k-2*marge); dy_vecteur.append(0)
                    couleur_vecteur.append(palette_vecteur[1])

                    if k>1: decalage1[k]+=0.05

                else:
                    couleur_point.append(palette_point[-2])
                    #vecteur pour p_rec(n,n)
                    x_debvecteur.append(n+decalage2); y_debvecteur.append(n+marge)
                    dx_vecteur.append(0); dy_vecteur.append(k-n-2*marge)
                    couleur_vecteur.append(palette_vecteur[2])

                    decalage2+=0.05

    #Ajout des vecteurs            
    plt.quiver(x_debvecteur, y_debvecteur, dx_vecteur, dy_vecteur,color=couleur_vecteur, angles='xy', scale_units='xy', scale=1, width=0.002)#, couleur_vecteur ) #palette_vecteur[2]
    #Ajout des points
    plt.scatter(x_point,y_point,color=couleur_point,s=50)           

    #Ajout des légendes
    plt.text(0, -1, 'Lecture du graphique :',fontsize = 20)

    place=-1.5
    for p in propriete_point:
        plt.text(0, place, 'Points vérifiant '+propriete_point[p], fontsize=16, color=palette_point[p])   
        place-=0.5
    for v in propriete_vecteur:
        plt.text(0, place, 'Vecteurs illustrant la dépendance de '+propriete_vecteur[v], fontsize=16, color=palette_vecteur[v])   
        place-=0.5

    #Affichage       
    plt.show()

plt.rcParams.update({'font.size': 14})    
Graphique_Lien_termes(7)    
```

On peut donc, au vu de ces dépendances, organiser le calcul des termes $p(n,k)$ en évitant de recalculer des termes et en évitant d'avoir à tester la redondance des calculs.

```python
def p_ite(n,k,res_mat=False):
    """
    Algorithme itératif de calcul des p(n,k)
    Par défaut la fonction renvoie p(n,k) 
    et l'argument res_mat permet de renvoyer la matrice complète des coefficients  
    """
    t=[ [0 for j in range(k+1)] for i in range(n+1) ]    
    #(initialisation de la ligne des 0 déjà effectuée)
    
    #initialisation de la colonne des 1
    for j in range(k+1): t[0][j]=1
    
    #parcours de toutes les colonnes
    for i in range(1,n+1):
        
        #remplissage au dessus de la diagonale
        for j in range(1,min(i+1,k+1)): #i>=j (+condition pour éviter débord du tableau)
            t[i][j]=t[i][j-1]+t[i-j][j]
        #remplissage en dessous de la diagonale
        for j in range(i+1,k+1): #i<j
            t[i][j]=t[i][i]
    
    return t[n][k] if not res_mat else t

Test_element(p_ite,30,18)
```

En négligeant naïvement le fait que le calcul d'un coefficient $p(N,K)$ avec cet algorithme permet en réalité d'obtenir toute la matrice des coefficients pour '$0 \leq n \leq N$ et $0 \leq k \leq K$', on peut effectuer une première comparaison de temps de calcul de coefficients pour quelques couples $(N,K)$ fixés.

```python
def Comp_temps(liste_fonc_p,liste_couples,txt="p"):
    """
    Affiche les temps de calcul des p(n,k) pour une liste de fonctions et une liste de couples (n,k)
    """   
    for couple in liste_couples:
        G={}
        for p in liste_fonc_p:
            G[p]=Test_temps_calcul(p,*couple)
        print("Pour le couple ",couple,", on a :",sep='')
        for p in liste_fonc_p:
            x,t=G[p]
            print("   Avec la fonction ",p.__name__,":")
            print("      ",txt,couple,"=",x," obtenu en ",t," s.",sep='')
        
    return None
      
Comp_temps([p_rec,p_rec_mem,p_ite],[(10,5),(20,7),(30,18),(40,25),(60,37),(70,62)])
```

Evidemment, il convient de retenir que le principal avantage de la fonction itérative, outre sa vitesse d'exécution, est qu'elle calcule en réalité l'ensemble des coefficients $p(n,k)$ pour tous les $0 \leq n \leq N$ et $0 \leq k \leq K$.

On peut réaliser (ci-dessous) le graphique donnant les $T(N,K)$, où $T(N,K)$ est le temps de calcul de la matrice des $p(n,k)$ pour $0 \leq n \leq N$ et $0 \leq k \leq K$. 


```python
plt.rcParams['figure.figsize'] = [20, 10] ; plt.rcParams.update({'font.size': 14})
Temps=Graphique_temps_fonction(p_ite,150,texte="\nGraphique donnant les temps T(N,K) de calcul des matrices des p(n,k) avec la fonction",legendex='N',legendey='K',legendez='T(N,K) (en s)')
```

<h2> <span style="font-size:0.7em; color:orange"> 3.1.1. Algorithme récursif direct </span> </h2> 




Dans <a href="#Titre21">la partie 2.1</a>, la recherche d'une formule de récurrence sur les $p(n,k)$ a déjà mis en évidence les liens qui existent entre les partitions des entiers, permettant une construction récursive des ensembles de partitions $P(n,k)$.

Rappelons que $P(n,k)$ est l'ensemble des partitions de $n$ de longueurs maximales $k$. On définit une partition de $n$ sous la forme d'un dictionnaire qui à une valeur apparaissant dans cette partition associe le nombre d'apparitions de cette valeur.

Par exemple, l'ensemble $P(5,3)=\left\{ \left(5\right);\left(1;4\right);\left(2;3\right);             \left(1;2;2\right);\left(1;1;3\right)\right\}$ sera codé sous la forme d'une liste d'objets (d'une classe partition), dont les contenus sont {5:1} , { 1:1 , 4:1 } , { 2:1 , 3:1 } , { 1:1 , 2:2 } et { 1:2 ; 3:1 }.


```python
class partition:
    """
    Classe des partitions
    """  
    
    def __init__(self,contenu={}):
        """
        Création d'une partition (vide par défaut)
        """
        self.contenu=contenu
       
    def __neg__(self):
        """
        Calcul de la longueur d'une partition
        Accès par la syntaxe préfixe -
        """
        return sum([self.contenu[k] for k in self.contenu])

    def __pow__(self,k):
        """
        Transformation d'une partition de P(n-k,k) en une partition de P(n,k) de longueur exactement k
        (transformation bijective)
        Accès par la syntaxe **
        """
        new={j+1:self.contenu[j] for j in self.contenu}
        new[1]=k-(-self)
        return partition(new)
        
    def __invert__(self):
        """
        Renvoie l'entier dont self est la partition
        Accès par la syntaxe ~
        """
        return sum([j*self.contenu[j] for j in self.contenu])
    
    def __str__(self):
        """
        Renvoie une chaîne de caractères décomposant la partition
        Conversion str
        """
        if not self.contenu: return "0=somme_vide"
        s=str(~self)+"="
        for k in self.contenu:
            for m in range(self.contenu[k]):
                s+=str(k)+"+"
        s=s[:len(s)-1]
        return s 
    
    def __repr__(self):
        """
        Représentation extérieure d'une partition
        Accès direct en console
        """
        return str(self)
```

```python
   
def P_rec(n,k):
    """
    Construction de la liste des partitions de n de longueurs inférieures ou égales à k
    """
    if n==0:
        return [partition()] #contient la partition vide
    if k==0:
        return [] #ne contient aucune partition
    if n>=k:
        res=P_rec(n,k-1)     
        for part in P_rec(n-k,k):                   
            res.append(part**k)
        return res
    return P_rec(n,n)
    
def Affichage_liste_partitions(P,n,k,end=' ; '):
    """
    Affichage des partitions de taille maximale k de l'entier n
    """
    print("\n\nLes partitions de longueur maximale ",k," de l'entier ",n," obtenues avec la fonction ",P.__name__," sont: \n",sep='')   
    nb_part=len(P(n,k))
    if nb_part==0: print("Aucune")
    else: 
        for part in P(n,k): print(part,end=end)
    return None


Affichage_liste_partitions(P_rec,15,15)        
```

<h2> <span style="font-size:0.7em; color:orange"> 3.1.2. Algorithme récursif avec mémorisation </span> </h2> 

De manière analogue à ce qui a été fait dans la <a href="#Titre222">partie 2.2.2</a>, on peut éviter la redondance de calculs de répartitions dans cette fonction, en mettant en place une matrice de mémorisation. 

```python
   
def P_rec_mem(n,k,mat=None):
    """
    Construction de la liste des partitions de n de longueurs maximales k
    """
    if not mat: mat=[[None for j in range(k+1)] for m in range(n+1)]
    
    if not mat[n][k]:
    
        if n==0:
            mat[n][k]=[partition()]
        elif k==0:
            mat[n][k]=[]
        elif n>=k: 
            res=P_rec(n,k-1)     
            for part in P_rec(n-k,k):                   
                res.append(part**k)
            mat[n][k]=res           
        else:
            mat[n][k]=mat[n][n]
        
    return mat[n][k]

        
Affichage_liste_partitions(P_rec_mem,15,15)   
```

<h2> <span style="font-size:0.7em; color:orange"> 3.1.3. Comparaison des algorithmes </span> </h2> 
  

De manière analogue à ce qui a été réalisé dans la <a href="#Titre223">partie 2.2.3.</a>, on peut réaliser un graphique permettant d'estimer le gain en temps d'exécution avec la matrice de mémorisation. 


Le graphique ci-dessous permet de visualiser les coefficients multiplicateurs exponentiels de gains en vitesse, calculés avec la formule $w(n,k)=ln \left( \frac{T[P-rec](n,k)}{T[P-rec-mem](n,k)} \right)$ où $T[P-rec](n,k)$ et $T[P-rec-mem](n,k)$ sont les temps de calcul des ensembles $P(n,k)$ respectivement avec les fonctions P_rec et P_rec_mem.

On remarque que les gains sont notables pour les couples $(n,k)$ tels que $ k \geq n$.

<h2> <span style="font-size:0.9em; color:orange"> 3.2. Algorithme itératif </span> </h2>  
    

En utilisant à nouveau les dépendances mises en évidence dans la <a href="#Titre23">partie 2.3.</a>, on peut éviter les tests en organisant le remplissage de la matrice des P(n,k) de façon optimale.


```python
def P_ite(n,k,res_mat=False):
    """
    Algorithme itératif de construction des ensembles P(n,k)
    Par défaut la fonction renvoie P(n,k) 
    et l'argument res_mat permet de renvoyer la matrice complète de ces ensembles  
    """
    t=[ [ [] for j in range(k+1)] for i in range(n+1) ]    
    #(initialisation de la ligne des 0 déjà effectuée)
    
    #initialisation de la colonne des 1
    for j in range(k+1): t[0][j]=[partition()]
        
    #parcours de toutes les colonnes
    for i in range(1,n+1):
                   
        #remplissage au dessus de la diagonale
        for j in range(1,min(i+1,k+1)): #i>=j (+condition pour éviter débord du tableau)
                
            res=t[i][j-1][:]
            
            for part in t[i-j][j]:
                res.append(part**j)  
            
            t[i][j]=res
            
        #remplissage en dessous de la diagonale
        for j in range(i+1,k+1): #i<j
            
            t[i][j]=t[i][i]

    return t[n][k] if not res_mat else t

Test_element(P_ite,15,15,txt="P")
```

Comme pour la fonction itérative p_ite vue dans la <a href="#Titre23"> partie 2.3 </a>, l'avantage de cette fonction itérative P_ite est de permettre le calcul de tous les ensembles $P(n,k)$ pour tous les $0 \leq n \leq N$ et $0 \leq k \leq K$.
